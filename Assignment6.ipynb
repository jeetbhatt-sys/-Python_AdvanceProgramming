```python
def find_the_difference(str1,str2):
    ref = ''
    for i in range(0,len(str1)):
        ref = str2
        for j in range(0,len(ref)):
            if str1[i] == ref[j]:
                if j > 0:
                    str2 = ref[0:j-1] + ref [j+1:]
                else:
                    str2 = ref[1:]
    return str2

str1 = input("Please Enter the first string : ")
str2 = input("Please Enter the second string : ")
print("Difference between the string : ", find_the_difference(str1,str2))
```

```python
def count_datatypes(*args):
    dict1 = {int : 0, str : 0, bool : 0, list : 0, tuple : 0, dict : 0}
    for element in args:
        dict1[type(element)] +=1
    return dict1.values()

print("Count of data types : ",count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23))
    
```

```python
def fib_str(n, lst):
    a = lst[0]
    b = lst[1]
    for i in range(2,n):
        t = b + a
        a = b
        b = t
        lst.append(t)
    return lst

n = int(input("Please Enter Value of n : "))
lst = input("Please enter the list elements comma seperated : ").split(',')
print("Fibonacci String : ", fib_str(n, lst))
```

```python
def ones_threes_nines(n):
    dict1 = { 'nines' : 0 , 'threes' : 0, 'ones' : 0}  
    if n / 9 > 0:
        dict1['nines'] = int(n/9)
        n -= dict1['nines'] * 9
    if n / 3 > 0:
        dict1['threes'] = int(n/3)
        n -= dict1['threes'] * 3
    if n / 1 > 0:
        dict1['ones'] = n
    return dict1
    
n = int(input("Please enter the value for n : "))
print("Output : ", ones_threes_nines(n))
```

```python
def fib(n):
    if n <= 1:
        return n
    else:
        res = int((1.61803398875**n - (-0.61803398875)**n)/2.236067977)
        return res

n = int(input("Please enter a integer : "))
print("Fibonacci Number at index {} : {}".format(n, fib(n)))
```
